```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
csv_path = "rollercoasters.csv"
rollercoaster_df = pd.read_csv(csv_path)
rollercoaster_df.tail()
```

```python
corrMatrix = rollercoaster_df.corr()
print (corrMatrix)
```

```python
import seaborn as sn
plt.figure(figsize = (16,5))
sn.heatmap(corrMatrix, annot=True)
fig = plt.show()
#highest drop height and avg speed .55
#ride length and excitement .56
#avg speed and max speed .66
#max psos gs and max sped .63 
#highest drop and max speed .93
#ride length and ride time .83
```

```python
lat_temp = plt.scatter(rollercoaster_df['excitement'], rollercoaster_df['intensity'], marker="o", facecolors="gold", edgecolors="black",
             alpha=0.75)

# # Incorporate the other graph properties

plt.xlabel("excitement")
plt.ylabel("intensity")

# # Incorporate a text label regarding circle size
plt.title("Excitement v Intensity")
# # Save Figure
plt.show()

#Botany Breakers is the outlier
```

```python
import seaborn as sns
sns.lmplot( x="excitement", y="intensity", data=rollercoaster_df, fit_reg=False, hue='custom_design', legend=False)
 
# Move the legend to an empty part of the plot
plt.legend(loc='upper right')
plt.title("Excitement v Intensity")
#Botany Breakers is the outlier
```
