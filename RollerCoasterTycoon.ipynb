```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
csv_path = "rollercoasters.csv"
rollercoaster_df = pd.read_csv(csv_path)
rollercoaster_df.head(10)
```

```python
corrMatrix = rollercoaster_df.corr()
print (corrMatrix)
```

```python
import seaborn as sn
sn.heatmap(corrMatrix, annot=True)
plt.show()
plt.savefig('test.jpg')
```

```python
fig = plt.gcf()
fig.set_size_inches(30, 30)
fig.savefig('test2png.png', dpi=100)
```

```python
lat_temp = plt.scatter(rollercoaster_df['excitement'], rollercoaster_df['intensity'], marker="o", facecolors="gold", edgecolors="black",
             alpha=0.75)

# # Incorporate the other graph properties

plt.xlabel("excitement")
plt.ylabel("intensity")

# # Incorporate a text label regarding circle size
plt.title("City Latitude v Temperature")
# # Save Figure
plt.show()
```
