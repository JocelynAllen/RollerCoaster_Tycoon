```python
#add mass of empty car, 100 kg (535 car+ 2* 100)
rollercoaster_df['Mass_Empty_Car_kg'] = '535'

#add the number of passengers per car
rollercoaster_df['Passengers_Car'] = '2'

#add the mass of the person (100)
rollercoaster_df['Mass_Person_kg'] = '100'
```

```python
#check if string or integer
rollercoaster_df.dtypes
```

```python
#convert objet to integer
rollercoaster_df['Mass_Empty_Car_kg'] = rollercoaster_df['Mass_Empty_Car_kg'].astype(int)
rollercoaster_df['Passengers_Car'] = rollercoaster_df['Passengers_Car'].astype(int)
rollercoaster_df['Mass_Person_kg'] = rollercoaster_df['Mass_Person_kg'].astype(int)
rollercoaster_df['Drop_feet'] = rollercoaster_df['Drop_feet'].astype(int)
```

```python
#the mass of the car (535 + (#of passengers/car * 100))
#mass of the car = (Mass_Empty_Car_kg + (Passengers_Car * Mass_Person_kg))
rollercoaster_df["Mass_Car_kg"] = rollercoaster_df["Mass_Empty_Car_kg"] +(rollercoaster_df["Passengers_Car"] * rollercoaster_df["Mass_Person_kg"])
```

```python
rollercoaster_df.head(10)
```

```python
#calculate the total mass per train; mass of the car * the number of cars per train
rollercoaster_df['Mass_Per_Train'] = ''

rollercoaster_df["Mass_Per_Train"] = rollercoaster_df["Mass_Car_kg"] * rollercoaster_df["cars_per_train"]

rollercoaster_df.head(10)
```

```python
#convert feet to meters
rollercoaster_df['Drop_ft_to_meters'] = ''

rollercoaster_df["Drop_ft_to_meters"] = rollercoaster_df["Drop_feet"] * .3048

rollercoaster_df.head(10)
```

```python
#Potential Energy Formula, PE = mgh, potential energy = mass* gravity (9.8)* height
rollercoaster_df['Potential_Energy'] = ''

rollercoaster_df["Potential_Energy"] = rollercoaster_df["Mass_Per_Train"] * 9.8 * rollercoaster_df["Mass_Per_Train"]

rollercoaster_df.head(10)
```
